```python
import pandas as pd
import re
```

### a.	Create the tracks.csv file by creating a new dataframe with only the following set of columns: [“artists”,“name”,“id”,“energy”,“popularity”,“year”]. You can get the year of the track from the given data.csv.

```python
data = pd.read_csv('./data/data.csv')
```

```python
# data.describe()
```

```python
data.head()
```

```python
tracks_keep_col_list = ['artists', 'name', 'id', 'energy', 'popularity', 'year']
```

```python
tracks_df = data[tracks_keep_col_list].copy(deep=True)
```

```python
tracks_df.head()
```

```python
tracks_df
```

```python
records = []
for idx, content in tracks_df.iterrows():
    if ',' in content['artists']:
        tmp = content['artists']
        artists_list = re.split('[,\'\"\[\]]', tmp)
        first_artist = 1
        for artist in artists_list:
            if artist != '' and artist != ' ':
                if first_artist:
                    tracks_df.at[idx, 'artists'] = artist
                    first_artist = 0
                else:
                    records.append([artist, content['name'], content['id'], \
                                    content['energy'], content['popularity'], content['year']])
    else:
        tmp = content['artists']
        artist = tmp[2:-2]
        tracks_df.at[idx, 'artists'] = artist
```

```python
len(records)
```

```python
tracks_df
```

```python
for rec in records:
    tracks_df.loc[len(tracks_df.index)] = rec
```

```python
tracks_df
```

```python
True in [*tracks_df.duplicated()]
```

```python
tracks_df_sorted = tracks_df.sort_values(by=['artists', 'name'])
```

```python
tracks_df_sorted
```

```python
tracks_df_sorted.to_csv('tracks.csv',index=False)
```

### b.	Create a artists.csv file without any duplicates with the following columns: [“artists”,“artist_popularity”,“energy”] from data_by_artists.csv. 
(Note: popularity given  in data_by_artists is artists popularity).

```python
data_artists = pd.read_csv('./data/data_by_artist.csv')
```

```python
data_artists.head()
```

```python
artists_keep_col_list = ['artists', 'popularity', 'energy']
```

```python
artists_df = data_artists[artists_keep_col_list].copy(deep=True)
```

```python
artists_df.head()
```

```python
artists_df
```

```python
artists_df.rename(columns={'popularity':'artist_popularity'}, inplace=True)
```

```python
artists_df
```

```python
artists_df.duplicated()
```

```python
True in [*artists_df.duplicated()]
```

```python
check_duplicated_artist = False
for status in artists_df.duplicated():
    check_duplicated_artist ^= status
check_duplicated_artist
```

```python
artists_df.to_csv('artists.csv',index=False)
```

### c.	Create a genres.csv file without any duplicates with the following columns: [“artists”,“genres”] from data_w_genre.csv.
