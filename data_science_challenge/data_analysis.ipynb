### a.	For every year in the data set, generate a table from completedata.csv of the most popular artist in each genre, descending from the year 2020 and save it as a .csv file.

```python
math.isnan(data['genres'][0])
```

```python
data_sort_by_year = data.sort_values(['year'], ascending=False, ignore_index=True)
data_sort_by_year = data_sort_by_year[data_sort_by_year['genres'] != "[]"]
```

```python
data_sort_by_year = data_sort_by_year.dropna(subset=['genres'])
```

```python
data_sort_by_year['genres'].isnull().sum()
```

```python
data_sort_by_year
```

```python
data_sort_by_year.loc[data_sort_by_year['year'] == 2021]
```

```python
popular_artists_data = []
for year in range(1920, 2021)[::-1]:
    print(year)
#     if year in data_sort_by_year['year']:
    popular_artists_data_popularity = {}
    popular_artists_data_artist = {}
    tmp_df = data_sort_by_year.loc[data_sort_by_year['year'] == year]
    for idx, row in tmp_df.iterrows():
        genres_str = row['genres']
        genres_list = re.split('[,\'\[\]]', genres_str)
        for genre in genres_list:
            if genre != '' and genre != ' ':
                if genre in popular_artists_data_popularity:
                    if row['popularity'] > popular_artists_data_popularity[genre]:
                        popular_artists_data_popularity[genre] = row['popularity']
                        popular_artists_data_artist[genre] = set()
                        popular_artists_data_artist[genre].add(row['artists'])
                    if row['popularity'] == popular_artists_data_popularity[genre]:
                        popular_artists_data_artist[genre].add(row['artists'])
                else:
                    popular_artists_data_popularity[genre] = row['popularity']
                    popular_artists_data_artist[genre] = set()
                    popular_artists_data_artist[genre].add(row['artists'])
    popular_artists_data.append(popular_artists_data_popularity)
    popular_artists_data.append(popular_artists_data_artist)
```

```python
popular_artists_data[0]
```

```python
popular_artists_data[1]
```

```python
len(popular_artists_data)
```

```python
popular_artists_2020_data = []
year = 2020
for i in range(len(popular_artists_data) // 2):
    for genre, popularity, artists_set in zip(popular_artists_data[i].keys(),\
                                  popular_artists_data[2*i].values(),\
                                  popular_artists_data[2*i+1].values()):
        tmp = []
        tmp.append(year)
        tmp.append(genre)
        tmp.append(artists_set)
        tmp.append(popularity)
        popular_artists_2020_data.append(tmp)
    year -= 1
```

```python
popular_artists_2020_data
```

```python
popular_artists_2020 = pd.DataFrame(popular_artists_2020_data, \
                                  columns=['year', 'genres', 'artists', 'popularity'])
```

### b.	Generate a table of the most popular track of the year and their popularity level from completedata.csv, and save it as a .csv file. If there’s a tie in popularity, list all tracks within that list.

```python
data_sort_by_year
```

```python
popular_tracks_data = {}
for year in range(1920, 2021)[::-1]:
    print(year)
    popular_tracks_data_popularity = None
    popular_tracks_data_track = set()
    tmp_df = data_sort_by_year.loc[data_sort_by_year['year'] == year]
    for idx, row in tmp_df.iterrows():
#         print(row['name'])
        if popular_tracks_data_popularity:
            if row['popularity'] > popular_tracks_data_popularity:
                popular_tracks_data_popularity = row['popularity']
                popular_tracks_data_track = set()
                popular_tracks_data_track.add(row['name'])
            if row['popularity'] == popular_tracks_data_popularity:
                popular_tracks_data_track.add(row['name'])
        else:
            popular_tracks_data_popularity = row['popularity']
            popular_tracks_data_track = set()
            popular_tracks_data_track.add(row['name'])
#     print(popular_tracks_data_track)
    popular_tracks_data[year] = [popular_tracks_data_popularity, popular_tracks_data_track] 
```

```python
popular_tracks_data
```

```python
popular_tracks_2020_data = []
for year, (popularity, tracks) in zip(popular_tracks_data.keys(), popular_tracks_data.values()):
    tmp = []
    tmp.append(year)
    tmp.append(popularity)
    tmp.append(tracks)
    popular_tracks_2020_data.append(tmp)
```

```python
popular_tracks_2020 = pd.DataFrame(popular_tracks_2020_data, \
                                  columns=['year', 'popularity', 'tracks'])
```
