### a.	From the generated tracks.csv, artists.csv, genres.csv, integrate to form new dataframe, and save into completedata.csv.
(Note: Please reduce duplicate artists, combine the genres of the artist into a list of values).

```python
data_tracks = pd.read_csv('./tracks.csv')
data_artists = pd.read_csv('./artists.csv')
data_genres = pd.read_csv('./genres.csv')
```

```python
data_tracks
```

```python
data_artists
```

```python
data_genres
```

```python
temp = pd.concat([data_artists, data_genres], axis=1, join='inner')
temp
```

```python
temp = temp.loc[:, ~temp.columns.duplicated()]
temp
```

```python
temp = temp.drop(['genre_popularity'], axis=1)
temp
```

```python
temp.rename(columns={'energy':'artists_energy'}, inplace=True)
temp
```

```python
completedata = data_tracks.join(temp.set_index('artists'), on='artists')
completedata
```

```python
completedata.to_csv('completedata.csv',index=False)
```

### b.	Given JSON file users/Vinjit.json. In the JSON field all_data/userdata/topartists, find all the artists which are present in the artists.csv into and save the dataframe with the following columns [“artists”,“popularity”,“id”,“username”] in user_artists_found.csv.
(Note: The given JSON file contains artists from regions other than the US. If you are not able to find them, fill them with N/A. Username can be found in json file).

```python
user_artists_found_data = []
for info in data_top_artists:
    tmp_data = []
    if info['name'] in artists_set:
        tmp_data.append(info['name'])
        tmp_data.append(data_artists.loc[data_artists['artists'] \
                                         == info['name']]['artist_popularity'].values[0])
        tmp_data.append(info['id'])
        tmp_data.append(username)
    else:
#         tmp_data.append('NaN')
        tmp_data.append(info['name'])
        tmp_data.append(info['popularity'])
        tmp_data.append(info['id'])
        tmp_data.append(username)
    print(tmp_data)
    user_artists_found_data.append(tmp_data)
```

```python
user_artists_found = pd.DataFrame(user_artists_found_data, \
                                  columns=['artists', 'popularity', 'id', 'username'])
user_artists_found
```

```python
user_artists_found.to_csv('user_artists_found.csv',index=False)
```
